```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
t_lim = 30 #límite de tiempo
gacela= []  #array para evolución temporal número de presas
leon = [] #array para evolución temporal número de depredadores
gc =50.0  #número inicial de presas
ln =100.0 #número inicial de depredadores
##constantes para rates
c1=1.0
c2=0.005
c3=0.6
currTime=0 #tiempo actual
```

```python
while(currTime < t_lim):
  if (ln>0 and gc>0): 
    rate1 = c1*gc
    rate2 = c2*gc*ln
    rate3 = c3*ln
    rate  = rate1+rate2+rate3
    Rtime = np.random.exponential(1/rate)
    D = np.random.multinomial(1,[rate1/rate,rate2/rate,rate3/rate])

    if(D [0]==1):
      gc+=1
    elif(D[1]==1):
      gc-=1
      ln+=1
    else: 
      ln-=1
    currTime = currTime+Rtime
    leon.append(ln)
    gacela.append(gc)

    if(gc>0 and ln<=0):
      rate = c1*gc
      Rtime = np.random.exponential(1/rate)
      gc+=1
      currTime = currTime+Rtime
      leon.append(ln)
      gacela.append(gc)
    
    if(ln<=0 and gc<=0): 
      currTime = t_lim 

    if(ln>0 and gc<=0):
      rate=c3*ln
      Rtime = np.random.exponential(1/rate)
      ln-=1
      currTime = currTime+Rtime
```

```python
plt.plot(leon, color="yellow")
plt.plot(gacela)
plt.title("Gacela/León ")
```
