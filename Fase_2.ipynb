# Actividad 1

```python
import math  
import matplotlib.pyplot as plt
import numpy as np

a,b,c,d,epsilon= 1,0.02,1,0.01,0.5

x0,y0,t0,tf,h=20,20,0,30,.01

nt = int((tf-t0)/h) #número de iteraciones

X= np.empty(nt+1)
Y= np.empty(nt+1)

X[0]=x0
Y[0]=y0
```

```python
for step in range(nt):
  X_1=a*X[step]-b*X[step]*Y[step]

  Y_1=-c*Y[step]+d*X[step]*Y[step]

  X[step+1] = X[step]-X_1*h
  
  Y[step+1] = Y[step]-Y_1*h
```

```python
fig, ax = plt.subplots()
plt.plot(X,'b-',label='Presas')
plt.plot(Y,'r-',label='Depredador')
ax.legend()

plt.title("Figura 1. Ciclicidad Presa/Depredador [O(h^2)] ")
```

```python
fig, ax = plt.subplots()
plt.plot(X,Y,'g-')
ax.legend()

plt.title("Figura 2. Órbitas Presa/Depredador [O(h^2)] ")
```

#Actividad 2

```python
for step in range(nt):
    X_1=a*X[step]-b*X[step]*Y[step]

    Y_1=-(c*Y[step])+d*X[step]*Y[step]

    X_2=a*X_1-b*((X_1*Y[step])+(X[step]*Y_1)) 
 
    Y_2=-(c*Y_1)+d*((X_1*Y[step])+(X[step]*Y_1))

    X[step+1] = X[step]+X_1*h+(1/2)*X_2*(h**2)

    Y[step+1] = Y[step]+Y_1*h+(1/2)*Y_2*(h**2)
```

```python
fig, ax = plt.subplots()
plt.plot(X,'b-',label='Presas')
plt.plot(Y,'r-',label='Depredador')
ax.legend()

plt.title("Figura 3. Ciclicidad Presa/Depredador [O(h^3)] ")
```

```python
fig, ax = plt.subplots()
plt.plot(X,Y,'g-')
ax.legend()

plt.title("Figura 4. Órbitas Presa/Depredador [O(h^3)] ")
```

#Actividad 3

```python
for step in range(nt):
  X_1=a*X[step]-b*X[step]*Y[step]-epsilon*X[step]

  Y_1=-c*Y[step]+d*X[step]*Y[step]-epsilon*Y[step]

  X[step+1] = X[step]-X_1*h
  
  Y[step+1] = Y[step]-Y_1*h
```

```python
fig, ax = plt.subplots()
plt.plot(X,'b-',label='Presas')
plt.plot(Y,'r-',label='Depredador')
ax.legend()

plt.title("Figura 5. Ciclicidad Presa/Depredador [O(h^2)] (epsilon=0.05")
```

```python
fig, ax = plt.subplots()
plt.plot(X,Y,'g-')
ax.legend()

plt.title("Figura 6. Órbitas Presa/Depredador [O(h^2)] (epsilon=0.05)")
```
